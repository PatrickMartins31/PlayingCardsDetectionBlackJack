```python
# Import opencv
import cv2

# Import uuid
import uuid

# Import Operating System
import os

# Import time
import time
```

```python
labels = ['AS', 'KS', '2S', '3S', '4S', '5S', '6S', '7S', '8S', '9S', '10S', 'QS', 'JS']
number_imgs = 5
```

```python
IMAGES_PATH = os.path.join('Tensorflow', 'workspace', 'images', 'collectedimages')
```

```python
if not os.path.exists(IMAGES_PATH):
    if os.name == 'posix':
        !mkdir - p {IMAGES_PATH}
    if os.name == 'nt':
        !mkdir {IMAGES_PATH}
for label in labels:
    path = os.path.join(IMAGES_PATH, label)
    if not os.path.exists(path):
        !mkdir {path}
```

```python
for label in labels:
    cap = cv2.VideoCapture(0)
    print('Collecting images for {}'.format(label))
    time.sleep(10)
    for imgnum in range(number_imgs):
        print('Collecting image {}'.format(imgnum))
        ret, frame = cap.read()
        imgname = os.path.join(IMAGES_PATH, label, label +
                               '.'+'{}.jpg'.format(str(uuid.uuid1())))
        cv2.imwrite(imgname, frame)
        cv2.imshow('frame', frame)
        time.sleep(5)
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2

cam = cv2.VideoCapture(0)
count = 0
while True:
    check, frame = cam.read()

    cv2.imshow('video', cv2.resize(
        frame, (1000, 800)))
    key = cv2.waitKey(1)
    if key == ord('1'):
        count += 1
        imgname = os.path.join(IMAGES_PATH, 'AS', 'AS' +
                               '.'+'{}.jpg'.format(str(uuid.uuid1())))
        cv2.imwrite(imgname, frame)
        print('screenshot taken', count)
    elif key == ord('2'):
        count += 1
        imgname = os.path.join(IMAGES_PATH, '2S', '2S' +
                               '.'+'{}.jpg'.format(str(uuid.uuid1())))
        cv2.imwrite(imgname, frame)
        print('screenshot taken 2S', count)
    elif key == ord('3'):
        count += 1
        imgname = os.path.join(IMAGES_PATH, '3S', '3S' +
                               '.'+'{}.jpg'.format(str(uuid.uuid1())))
        cv2.imwrite(imgname, frame)
        print('screenshot taken 3S', count)
    elif key == ord('4'):
        count += 1
        imgname = os.path.join(IMAGES_PATH, '4S', '4S' +
                               '.'+'{}.jpg'.format(str(uuid.uuid1())))
        cv2.imwrite(imgname, frame)
        print('screenshot taken 4S', count)
    elif key == ord('5'):
        count += 1
        imgname = os.path.join(IMAGES_PATH, '5S', '5S' +
                               '.'+'{}.jpg'.format(str(uuid.uuid1())))
        cv2.imwrite(imgname, frame)
        print('screenshot taken 5S', count)

    elif key == ord('6'):
        count += 1
        imgname = os.path.join(IMAGES_PATH, '6S', '6S' +
                               '.'+'{}.jpg'.format(str(uuid.uuid1())))
        cv2.imwrite(imgname, frame)
        print('screenshot taken 6S', count)
    elif key == ord('7'):
        count += 1
        imgname = os.path.join(IMAGES_PATH, '7S', '7S' +
                               '.'+'{}.jpg'.format(str(uuid.uuid1())))
        cv2.imwrite(imgname, frame)
        print('screenshot taken 7S', count)
    elif key == ord('8'):
        count += 1
        imgname = os.path.join(IMAGES_PATH, '8S', '8S' +
                               '.'+'{}.jpg'.format(str(uuid.uuid1())))
        cv2.imwrite(imgname, frame)
        print('screenshot taken 8S', count)
    elif key == ord('9'):
        count += 1
        imgname = os.path.join(IMAGES_PATH, '9S', '9S' +
                               '.'+'{}.jpg'.format(str(uuid.uuid1())))
        cv2.imwrite(imgname, frame)
        print('screenshot taken 9S', count)
    elif key == ord('0'):
        count += 1
        imgname = os.path.join(IMAGES_PATH, '10S', '10S' +
                               '.'+'{}.jpg'.format(str(uuid.uuid1())))
        cv2.imwrite(imgname, frame)
        print('screenshot taken 10S', count)
    elif key == ord('k'):
        count += 1
        imgname = os.path.join(IMAGES_PATH, 'KS', 'KS' +
                               '.'+'{}.jpg'.format(str(uuid.uuid1())))
        cv2.imwrite(imgname, frame)
        print('screenshot taken KS', count)
    elif key == ord('q'):
        count += 1
        imgname = os.path.join(IMAGES_PATH, 'QS', 'QS' +
                               '.'+'{}.jpg'.format(str(uuid.uuid1())))
        cv2.imwrite(imgname, frame)
        print('screenshot taken QS', count)
    elif key == ord('j'):
        count += 1
        imgname = os.path.join(IMAGES_PATH, 'JS', 'JS' +
                               '.'+'{}.jpg'.format(str(uuid.uuid1())))
        cv2.imwrite(imgname, frame)
        print('screenshot taken JS', count)
    elif key == ord('b'):
        cam.release()
        cv2.destroyAllWindows()
        break
```
